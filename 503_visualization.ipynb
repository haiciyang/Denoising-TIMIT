```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sgd_err = np.load('SGD2tr_err.npy')
adam_err = np.load('Adam2tr_err.npy')
adag_err = np.load('Adagrad2tr_err.npy')
adab_err = np.load('AdaBound2tr_err.npy')

sgd_tr_sdr = np.load('SGD2tr_sdr.npy')
adam_tr_sdr = np.load('Adam2tr_sdr.npy')
adag_tr_sdr = np.load('Adagrad2tr_sdr.npy')
adab_tr_sdr = np.load('AdaBound2tr_sdr.npy')

sgd_te_sdr = np.load('SGD2te_sdr.npy')
adam_te_sdr = np.load('Adam2te_sdr.npy')
adag_te_sdr = np.load('Adagrad2te_sdr.npy')
adab_te_sdr = np.load('AdaBound2te_sdr.npy')
```

```python
N_iter = 50
```

```python
# plt.plot(range(N_iter), [sgd_err, adam_err, adag_err, adab_err])
```
